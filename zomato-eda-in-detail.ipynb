 # Cleaning rate Column

```python
df['rate'].unique()
```

Lets remove "Nan","-" and "/5" for all data from row column

```python
#Creating user defined Function
def handlerate(value):
    if(value=='NEW' or value=='-'):   # here if the values are considered to nan or replaced as nan
        return np.nan   
    else:
        value=str(value).split('/')     # this splits the value into 2 different list 
        value=value[0]                  # from this splited list we consider the 1st value or 0 position value
        return float(value)
    
df['rate']=df['rate'].apply(handlerate)  # we can also use function handlerate(df['rate'])
df['rate'].head()

# now we have nan values so lets remove it.
```

```python
df['rate'].isnull().sum()
```

Filling Null Values in Rate Column with Mean 

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info() 
#Here we can see that rate is normalized or has equal no of rows in the column
```

# So Let's remove comma or ',' from the values to make it either integer or float from string
# now lets create a new user defined function for performing the removal or comma

```python
# removing " , " from cost2plate column 
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)

df['Cost2plates']=df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

## Cleaning Rest Type Column

```python
#Lets see the Value counts 
df['rest_type'].value_counts()
```

```python
#Here Lets cluster the values as others for restorents less than 1000 and make it a cluster or group them together as another category.
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types #This is a dupliucate of the df[rest_type] as rest_type.
```

```python
# now Clustering restorent less than 1000 as one cluster
rest_type_lessthan1000=rest_types[rest_types<1000]
rest_type_lessthan1000
```

## Making Rest Types less than 1000 in frequency as others

```python
# Naming the Restorent which are less than 1000 as Others this makes the data clean and easy to analysis.
# user define function for renaming the values less than 1000
def handle_rest_type(value):
    if(value in rest_type_lessthan1000):  # which ever values are presnt in rest_type and are in rest_type_lessthan1000 or values <1000, will be renames as others.
        return 'others'
    else:
        return value

df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

Here we can see that the Values others contain less than 1000 or 9003

## Cleaning Location Column

```python
location = df['location'].value_counts(ascending  = False) # arranging location in desending order

location_lessthan300 = location[location<300]  # location that aare less than 3000



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

# Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
